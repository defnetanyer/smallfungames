```python
# finds the root of whatever
def root(num, root):
    try:
        result = num ** (1/root)
        return result
    except ZeroDivisionError: 
        return "Cannot root by 0 (zero)"
   
print(root(4,2))
print(root(5,0))
print(root(4,3))
```

```python
#Rock Paper Scissors IT WORKS
import random
robot_won= "I have bested you! HAHA "
player_won= "You have won, you are very smart. "
tie = "We have appeared to have tied! "

def ropasc():
    options = ['rock', 'paper', 'scissors']
    num=random.randint(0,2)
    robot = options[num]
    player = input("Rock, Paper, Scissors, Shoot! ")
    result = None
    
    if player == robot:
        return tie
    if player == "rock" and robot == "paper":
        return robot_won
    if player == "rock" and robot == "scissors":
        return player_won
    if player == "paper" and robot == "rock":
        return player_won
    if player == "paper" and robot == "scissors":
        return robot_won
    if player == "scissors" and robot == "rock":
        return robot_won
    if player == "scissors" and robot == "paper":
        return player_won

def main():
    running = True
    while running == True:
        #one game
        inp = input("Want to play rock, paper, scissors? Press enter to continue, press T for a tournaments and press any key to stop! ")
        if inp == '':
            #ropasc()
            print(ropasc())
            now_what = input("Would you like to stop playing (any key) or return to main menu? ")
            if now_what == "return to main" or now_what == "main":
                main()
            else:
                running = False
        
        #tournament!!
        if inp == 't' or inp == 'T':
            games = int(input("How many rounds would you like to play? "))
            rounds = 0
            rScore = 0
            pScore = 0
            for rounds in range (0, games):
                result = ropasc()
                if result == robot_won:
                    rScore = rScore +1
                    rounds = rounds+1

                if result == player_won:
                    pScore = pScore +1
                    rounds = rounds+1

                if result == tie:
                    rounds = rounds +1

            if rScore > pScore:
                print(robot_won)
            if pScore > rScore:
                print(player_won)
            if pScore == rScore:
                print(tie)

            now_what2 = input("Would you like to stop (press any key) or go to main menu?" )
            if now_what2 == "return to main" or now_what2 == "main":
                main()

            else:
                running = False

main()
```
